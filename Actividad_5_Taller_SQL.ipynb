```python
#6. Para cada ProductID, suma la cantidad vendida (SUM(OrderQty)). Devuelve solo productos con ventas totales > 1000,
#ordenados de manera descendente.Usa GROUP BY, HAVING y ORDER BY
respuesta6_query = ("""
SELECT p.ProductID, (SUM(o.OrderQty)) AS TotalQuantity
FROM SalesLT.Product AS p
INNER JOIN SalesLT.SalesOrderDetail AS o ON p.ProductID = o.ProductID
GROUP BY p.ProductID
HAVING SUM(o.OrderQty) > 1000
ORDER BY TotalQuantity DESC;
""")
#Selecting SProductID from SalesLT.Product table
#Defining (SUM(of.Order Qty)) as TotalQuantity
#o is also known a SalesOrderDetail
#p is also known a SalesLT.Product
#Joining the information ProductID of SalesOrderDetail table and SalesLT.Product table when the information coincide
#Grouping the matching information from p.ProductID = o.ProductID
#Filtering total sales > 1000
#Order price in descending order
respuesta6 = pd.read_sql_query(respuesta6_query, conn)
display(respuesta6)
print("There arent any products with a total quantity greater than 1000")

```

```python
#7. Productos sin ventas: lista productos que no aparecen en SalesLT.SalesOrderDetail.
#Usa LEFT JOIN + WHERE
respuesta7_query = ("""
SELECT p.ProductID, p.Name AS ProductName
FROM SalesLT.Product AS p
LEFT JOIN SalesLT.SalesOrderDetail AS o ON p.ProductID = o.ProductID
WHERE o.ProductID IS NULL;
""")
#Selecting p.ProductID, p.Name from SalesLT.Product table
#Joining the ALL information ProductID of SalesOrderDetail table and SalesLT.Product table
#Show only products that are not selling.
respuesta7 = pd.read_sql_query(respuesta7_query, conn)
display(respuesta7)
```

```python
#8. Clientes por país: total de clientes por CountryRegion (usa Address + CustomerAddress).
#Podrias necesitar: COUNT(DISTINCT CustomerID).
respuesta8_query = ("""
SELECT a.CountryRegion, COUNT(DISTINCT c.CustomerID) AS TotalCustomers
FROM SalesLT.Address AS a
INNER JOIN SalesLT.CustomerAddress AS ca ON a.AddressID = ca.AddressID
INNER JOIN SalesLT.Customer AS c ON ca.CustomerID = c.CustomerID
GROUP BY a.CountryRegion;
""")
#Selecting a.CountryRegion FROM SalesLT.Address table
respuesta8 = pd.read_sql_query(respuesta8_query, conn)
display(respuesta8)
```
